```python
''' Takes in dataframe to change data type to datetime and returns the count per year
'''
def countPerYear(df):
    df['Date'] = pd.to_datetime(df['Date'])
    return df['Date'].dt.year.value_counts()
```

```python
def changeToDateTime(df, columnName, timeFormat): 
    df[columnName] = pd.to_datetime(df[columnName], format = timeFormat)
```
