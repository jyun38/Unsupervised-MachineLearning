```python
data_dropDate = data_addHour.drop(['Date'], axis=1)
```

```python
finalDf = data_dropDate
finalDf
```

```python
year2002 = finalDf[finalDf.Year== 2002]
year2002
```

```python
def num_clusters(dataset):
    num_observations = len(dataset)
    value = math.sqrt(num_observations/2)
    return value
    
```

```python
km = KModes(n_clusters=495, init = "Huang", n_init = 3, verbose=1)
```

```python
clusters = km.fit_predict(year2002)
```

```python
cluster_centroids = pd.DataFrame(km.cluster_centroids_)
cluster_centroids
```
